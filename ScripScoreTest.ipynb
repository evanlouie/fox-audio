```python
# Find the hits inside the new df for gunshots
# for loop that goes through one value of LabelConfidencePairs and creates a new column based on label[n] and labelConfidence[]
import ast

f4 = f3
for index, row in f3.iterrows():
    rowStr = row['LabelConfidencePairs']
    rowLabels = rowStr.split()
    if(index == 0):
        for newColumn in range(0, int(len(rowLabels)/2)):
            iterativeLabel = "label_"+str(newColumn)
            iterativeLabelConfidence = "labelConfidenceRate_"+str(newColumn)
            f4 = f4.reindex(columns=[*f4.columns.tolist(), iterativeLabel, iterativeLabelConfidence], fill_value=0.0)
    
    for newVal in range(0, int(len(rowLabels))): 
            #print (str(index) + "and" + str(newVal) + "and" + rowLabels[newVal])
            
            #insertLabelConfidence = "labelConfidenceRate_"+str(0)
            
            if newVal % 2 == 0:
                if(newVal <=1):
                    x=0
                else:
                    x = newVal/2
            
            insertLabel = str("label_"+str(int(x)))
            insertLabelConfidence = "labelConfidenceRate_"+str(int(x))
            
            # write to dataframe
            
                
            #print(insertLabel + "   " + insertLabelConfidence)
            if newVal % 2 == 0:
                f4.at[index, insertLabel] = rowLabels[newVal]
            
            else:
                #f4.set_value(index, insertLabelConfidence, float(rowLabels[newVal]))
                f4.at[index, insertLabelConfidence] = ast.literal_eval(rowLabels[newVal])
                #f4.at[index, insertLabelConfidence] = float(10.910)
            #else
            #    f4.insert(column=insertLabelConfidence)
                
            
            
falsePos = pd.concat([f1,])
for indexfp, row2 in f1.iterrows():
    rowStrfp = row2['LabelConfidencePairs']
    rowLabelsfp = rowStrfp.split()
    if(indexfp == 0):
        for newColumn2 in range(0, int(len(rowLabelsfp)/2)):
            iterativeLabelfp = "label_"+str(newColumn2)
            iterativeLabelConfidencefp = "labelConfidenceRate_"+str(newColumn2)
            falsePos = falsePos.reindex(columns=[*falsePos.columns.tolist(), iterativeLabelfp, iterativeLabelConfidencefp], fill_value=0.0)
    
    for newValfp in range(0, int(len(rowLabelsfp))): 
            #print (str(index) + "and" + str(newVal) + "and" + rowLabels[newVal])
            
            #insertLabelConfidence = "labelConfidenceRate_"+str(0)
            
            if newValfp % 2 == 0:
                if(newValfp <=1):
                    x=0
                else:
                    x = newValfp/2
            
            insertLabelfp = str("label_"+str(int(x)))
            insertLabelConfidencefp = "labelConfidenceRate_"+str(int(x))
            
            # write to dataframe
            
                
            #print(insertLabel + "   " + insertLabelConfidence)
            if newValfp % 2 == 0:
                falsePos.at[indexfp, insertLabelfp] = rowLabelsfp[newValfp]
            
            else:
                #f4.set_value(index, insertLabelConfidence, float(rowLabels[newVal]))
                falsePos.at[indexfp, insertLabelConfidencefp] = ast.literal_eval(rowLabelsfp[newValfp])
                #f4.at[index, insertLabelConfidence] = float(10.910)
            #else
            #    f4.insert(column=insertLabelConfidence)



            
```
